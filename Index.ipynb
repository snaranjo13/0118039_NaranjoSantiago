```python
#%% SÉPTIMO EJERCICIO
lista = list(input('Dígite una lista de numeros en orden ascendente:'))
lista_ord = lista.copy() # Crea una copia para ser mutada
lista_ord.sort() # Ordena las listas de menor a mayor

if lista == lista_ord: # Determina si la lista está en orden ascendente
    print('La lista está ordenada de forma ascendente')
    a = int(input('Ahora, dígite un número:')) 
    for i in range(0,len(lista)+1):
        if i >= a: # Este if determina la posición donde debe ingresarse a 
            aux1 = lista[:i] # Toma todos los elementos hasta i 
            aux2 = lista[i:] # Toma todos los elemnetos posteriores a i 
            lista_nueva = aux1+[a]+aux2
            print('La nueva lista es:',lista_nueva)
            break
            
else:
    print('La lista no está ordenada de forma ascendente')
```

```python
#%% OCTAVO EJERCICIO
lista = eval(input('Dígite una lista:'))
maxi = max(lista)
nueva_l = []

for i in range(0,len(lista)):
    if lista[i] < maxi: 
        nueva_l.append(lista[i]) # Toma los elementos menores al primer máximo 
                                 # y los guarda en un nuevo vector
print ('El segundo número mayor es: ',max(nueva_l))
```

```python
#%% DÉCIMO EJERCICIO
# NOTA: Este fue el único ejercicio que no fui capaz de hacer, además por tema
# de tiempo no pude pedir monitoria, así que hice mi mayor esfuerzo por 
# entender el código de un compañero. De antemano gracias por la comprensión.

M = []

filas = int(input('Dígite el número de filas de la matriz: '))
columnas = int(input('Dígite el número de columnas de la matriz: '))    

for i in range(filas): # Este for crea la matriz pidiendo al usuario cada
                       # elemento
    M.append([])
    
    for j in range(columnas):
        M[i].append(int(input(F'Escriba el elemento #{j+1} de la fila {i+1}: '\
         )))        

num_elementos = filas*columnas # Cantidad de elementos de la matriz
vec = []
col_final = columnas - 1
col_inicial = 0
fil_final = filas - 1
fil_inicial = 0
cont = 0

while cont < num_elementos: 
    
    for i in range(col_final,col_inicial,-1): # Recorrido izquierda-derecha
        pos = M[fil_final][i]
        vec.append(pos)
        cont += 1
        
    for i in range(fil_final,fil_inicial,-1): # Recorrido abajo-arriba
        pos = M[i][col_inicial]
        vec.append(pos)
        cont += 1
    
    for i in range(col_inicial,col_final): # Recorrido derecha-izquierda
        pos = M[fil_inicial][i]
        vec.append(pos)
        cont += 1
    
    for i in range(fil_inicial,fil_final): # Recorrido arriba abajo
        pos = M[i][col_final]
        vec.append(pos)
        cont += 1
    
    col_final -= 1
    col_inicial += 1
    fil_final -= 1
    fil_inicial += 1
    # Se redefinen para cambiar de posiciones inicales
print('El vector solicitado es: ',vec)    
```

```python
#%% UNDÉCIMO EJERCICIO
ancho = int(input("Dígite el ancho del rectángulo: "))
altura = int(input("Altura del rectángulo: "))
caracter = str(input('Dígite el caracter a utilizar:'))

def dibujo(ancho,altura,caracter):
   
    for i in range(altura): # Recorre horizontalmente
        for j in range(ancho): # Recorre verticalmente
            print(caracter, end="") # El 'end=""' es para que no añada un salto
                                    # entre línea            
        print()

dibujo(ancho,altura,caracter)
```

```python
#%% DECIMOSEGUNDO EJERCICIO
ancho = int(input('Dígite el ancho del triángulo:'))

def dibujo(ancho):
    
    for i in range(ancho+1): # Recorre la primera mitad del triángulo
        print("*"*i) # El '*i' sirve para hacer el dibujo, si no estuviese
                     # haría una línea recta
    for j in range(ancho-1,0,-1): # Recorre la otra mitad
        print("*"*j)
        
dibujo(ancho)
```

```python
#%% DECIMOTERCER EJERCICIO
año1 = int(input('Dígite un año:'))
año2 = int(input('Dígite otro año:'))

def bisiestos(año1,año2):
    
    años_bisiestos = 0

    if año1 <= año2: # Este if permite al usuario no tener que escribir los 
                     # años de forma ascendente
        for i in range(año1,año2):
            if i%4 ==0: # Este if determina si es bisiesto
                años_bisiestos = años_bisiestos + 1
        diferencia = año2 - (año1 - 1)
        print('De ',año1,' a ',año2,' hay ',diferencia,' años, '\
              ,años_bisiestos,' son bisiestos.')
    else:
        for i in range(año2,año1):
            if i%4 ==0:
                años_bisiestos = años_bisiestos + 1 
        diferencia = año1 - (año2 - 1)
        print('De ',año2,' a ',año1,' hay ',diferencia,' años, '\
              ,años_bisiestos,' son bisiestos.')
        
        
bisiestos(año1,año2)
```

```python
#%% DECIMOCUARTO EJERCICIO
def suma():
    
    print('Escriba el resultado de las siguientes operaciones:')
    aciertos = 0
    
    while aciertos < 5: # Se mantiene en el ciclo hasta que complete los 5 
                        # aciertos
        a = random.randint(1,100) # Crea un número aleatorio
        b = random.randint(1,100)
        c = a + b
        print('\n') 
        print(a,'+',b)
        d = int(input('El resultado es:'))
        
        if d == c: # Compara si la suma es correcta
            print('El resultado es correcto.')
            aciertos = aciertos +1
        else:
            print('El resultado es incorrecto.')
            
    print('Programa terminado.')
    
suma()            
```

```python
#%% DECIMOQUINTO EJERCICIO


def MasterMin():

    cadena = []
    lon = int(input('Dígite la longitud de la cadena:'))
    for x in range(0,lon): # Esto es para crear una cadena aleatoria
        cadena.append(random.randint(0,9))
    print(cadena)

    acierto = 0
    while acierto != lon:
        acierto = 0
        cadena_usuario = eval(input('Intente adivinar la cadena:'))
        for i in range(0,lon):
            if cadena[i] == cadena_usuario[i]: # Compara si el elemneto de la
                                               # cadena es igual a lo digitado
                                               # por el usuario
                acierto = acierto + 1
        if acierto == 0:
            print('No ha acertado algún valor.')
        elif acierto < lon:    
            print('Con',cadena_usuario,'ha acertado',acierto,'valores.')
                   
    print('Con',cadena_usuario,'ha adivinado los',lon,'valores.')
  
MasterMin()        
```
